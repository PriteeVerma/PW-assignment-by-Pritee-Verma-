 **Question 1: What is the difference between multithreading and multiprocessing?**
-

**Answer :**
-
- Multithreading and multiprocessing are both ways to make a program run multiple tasks at the same time, but they work differently.

- Multithreading means running multiple threads inside a single process. Threads share the same memory space, so it’s good for tasks that are not CPU-heavy, like downloading files, reading data, or user interactions. But in Python, because of the Global Interpreter Lock (GIL), only one thread runs Python code at a time, so it’s not perfect for CPU-heavy work.

- Multiprocessing, on the other hand, runs multiple processes, each with its own memory and Python interpreter. It’s great for CPU-intensive work like calculations, image processing, or big data analysis because it truly runs tasks in parallel using multiple CPU cores.

- In short: multithreading is good for I/O tasks (like reading files or network calls) while multiprocessing is better for CPU-bound tasks.

**Question 2: What are the challenges associated with memory management in Python?**
-

**Answer:**
-
- Python manages memory automatically using garbage collection, but sometimes this causes challenges.

**Some common issues are:**
-

- Garbage collection overhead: Automatic cleanup can slow down performance if there are many unused objects.

- Reference cycles: If two objects refer to each other, Python’s garbage collector may not delete them properly, wasting memory.

- Memory leaks: When objects are not freed due to references or improper coding, memory keeps getting filled.

- Handling large data: Programs that deal with very large datasets (like in Pandas or NumPy) can quickly use up available memory.

- Fragmentation: Frequent creation and deletion of objects can fragment memory and reduce efficiency.

- To handle these, developers should write clean code, use efficient data structures, and free unused variables when possible.

**Question 3: Write a Python program that logs an error message to a log file when a division by zero exception occurs.**
-
**Answer:**
-

```python
import logging

# Set up logging
logging.basicConfig(filename='error.log', level=logging.ERROR)

try:
    a = 10
    b = 0
    result = a / b
except ZeroDivisionError as e:
    logging.error("Division by zero occurred: %s", e)
    print("Error logged successfully.")
```

**Explanation:**
-
- Here, if a division by zero happens, instead of crashing, the error is recorded in a file named error.log. This helps in debugging later without losing information.

**Question 4: Write a Python program that reads from one file and writes its content to another file.**
-

**Answer:**
-

```python
try:
    with open('source.txt', 'r') as source_file:
        data = source_file.read()

    with open('destination.txt', 'w') as dest_file:
        dest_file.write(data)

    print("File content copied successfully!")
except FileNotFoundError:
    print("The source file does not exist.")
```

**Explanation:**
-
- This program reads data from source.txt and writes it into destination.txt. If the source file doesn’t exist, it shows an error message instead of stopping.

**Question 5: Write a program that handles both IndexError and KeyError using a try-except block.**
-

**Answer:**
-

```python
my_list = [10, 20, 30]
my_dict = {'a': 1, 'b': 2}

try:
    print(my_list[5])     # This will cause IndexError
    print(my_dict['z'])   # This will cause KeyError
except IndexError:
    print("IndexError: You tried to access a list index that doesn't exist.")
except KeyError:
    print("KeyError: The key you tried to access is not in the dictionary.")
```

**Explanation:**
-
- This code safely handles two common errors — one from lists (IndexError) and one from dictionaries (KeyError) — so that the program doesn’t stop suddenly.

**Question 6: What are the differences between NumPy arrays and Python lists?**
-

**Answer:**
-
- Both NumPy arrays and Python lists are used to store collections of data, but they are quite different in how they work and how fast they perform.

**Data Type:**
-

- Python lists can store different types of data together, like numbers, strings, and objects in the same list.

- NumPy arrays can store only one type of data (for example, all integers or all floats).

**Speed and Performance:**
-

- NumPy arrays are much faster because they are implemented in C language and support vectorized operations.

- Python lists are slower since they store elements as separate objects.

**Memory Usage:**
-
- NumPy arrays use less memory and are more efficient for large data.

- Lists take up more memory as they store additional information about each element.

**Mathematical Operations:**
-

- NumPy arrays allow direct mathematical operations like addition or multiplication on all elements at once.

- In lists, we need to use loops to perform such operations manually.

**Functionality:**
-

- NumPy arrays have many built-in mathematical and statistical functions.

- Python lists are more general-purpose and don’t support these operations directly.

**Example:**



```python
import numpy as np

#Python list
list1 = [1, 2, 3]

#NumPy array
arr = np.array([1, 2, 3])

print(list1 * 2)  # [1, 2, 3, 1, 2, 3]
print(arr * 2)    # [2 4 6]
```

 **Explanation:**
-
- In this example, multiplying a list repeats its elements, while multiplying a NumPy array multiplies each value.
- That’s why NumPy arrays are mostly used in data science and numerical computations, while Python lists are better for general use.

**Question 7: Explain the difference between apply() and map() in Pandas.**
-

**Answer:**
-
- Both apply() and map() are used to perform operations on data in Pandas, but they are slightly different.

- map(): Used with a single Series (one column). It applies a function to each element.

- apply(): Used with both Series and DataFrame. It can apply a function to rows or columns.

**Example:**

```python
import pandas as pd

s = pd.Series([1, 2, 3])
print(s.map(lambda x: x * 2))

df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
print(df.apply(sum))
```

**In short :**
-
- map() is simpler and used for one column, while apply() is more flexible for multiple columns or custom operations.

**Question 8: Create a histogram using Seaborn to visualize a distribution.**
-
**Answer :**
-
- A histogram is a type of graph that shows how data is distributed — for example, how many times a value appears in a dataset. In Python, we can easily create a histogram using the Seaborn library.

**Here’s a simple example:**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
data = sns.load_dataset('iris')

# Create a histogram for sepal length
sns.histplot(data['sepal_length'], bins=20, kde=True, color='skyblue')

# Add title and labels
plt.title('Distribution of Sepal Length in Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

**Explanation:**
-
- We first import Seaborn and Matplotlib to make the graph.

- The iris dataset is a sample dataset that comes with Seaborn.

- sns.histplot() draws the histogram. The bars show how many times each sepal length value appears.

- kde=True adds a smooth line to show the overall shape of the data.

- Finally, plt.show() displays the graph.

This graph helps us easily see which sepal lengths are common and how the data is spread.

**Question 10: Calculate the correlation matrix using Seaborn and visualize it with a heatmap.**
-
**Answer:**
-
- A correlation matrix shows how much the columns of a dataset are related to each other. The values range from -1 to 1:

- 1 means two columns are strongly positively related.

- -1 means they are strongly negatively related.

- 0 means no relation.

- We can use Seaborn to make a heatmap for a visual view.

**Example:**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
df = sns.load_dataset('iris')

# Calculate correlation matrix
corr = df.corr()

# Create a heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

**Explanation:**
-
- This program calculates how strongly different columns are related to each other in a dataset. The heatmap uses colors to show positive or negative correlations — helpful for data analysis and feature selection.

**Summary:**
-
- This is useful in data analysis because we can understand relationships between columns and make better decisions when analyzing or building models.
