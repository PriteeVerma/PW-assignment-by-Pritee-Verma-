**Question 1: What is Simple Linear Regression?**
-

**Answer :**
-


Simple Linear Regression is a statistical technique used to study the relationship between one independent variable (X) and one dependent variable (Y). The goal is to fit a straight line that best represents the pattern in the data. This line is defined by the equation:

  Y=b0​+b1​X
  

where b₀ is the intercept and b₁ is the slope.
Simple Linear Regression helps in predicting the value of Y for any given value of X and also explains how strongly the two variables are related.

**Key Characteristics:**
-
● Uses only one independent variable.

● Assumes a linear and continuous relationship between X and Y.

● Minimizes error using the least-squares method.

● Useful for forecasting and understanding trends.

**Example:**
-
- Predicting house prices based on only the size (in sq. ft.) of the house.

**Question 2: What are the key assumptions of Simple Linear Regression?**
-

**Answer:**
-


Simple Linear Regression relies on several assumptions to produce accurate and reliable results. These assumptions ensure that the model generalizes well and that its statistical conclusions are valid.

**Key Assumptions:**
-

**1. Linearity:**

 The relationship between X and Y must be linear. If the relationship is curved or irregular, the model will fail to capture the pattern.

**2. Independence of Observations:**

Each data point should be independent of others. Dependencies create bias in model outcomes.

**3. Homoscedasticity (Constant Variance):**

The residuals (errors) should have the same spread across all values of X. Unequal variance leads to biased estimates.

**4. Normal Distribution of Residuals:**

The residuals should follow a normal distribution to ensure reliable hypothesis testing.

**5. No Major Outliers:**

Outliers can distort the regression line and reduce model performance.

**Importance:**
-

When these assumptions hold, the regression model becomes more interpretable, stable, and accurate.

**Question 3: What is heteroscedasticity, and why is it important to address in regression models?**
-
**Answer:**
-

Heteroscedasticity refers to a condition in which the residuals of a regression model do not have constant variance. Instead, the spread of errors increases or decreases as X changes. This violates one of the key assumptions of linear regression.

**Key Characteristics:**
-

● Residuals display an increasing or decreasing pattern.

● Often visible as a funnel shape in residual plots.

● Indicates that the model may be missing important variables or relationships.


**Why It Matters:**

● Leads to unreliable standard errors.

● Affects confidence intervals and hypothesis testing.

● Reduces the stability and accuracy of predictions.

Addressing heteroscedasticity helps ensure that the regression model is fair, unbiased, and mathematically valid.

**Question 6: Implement a Python program to fit a Simple Linear Regression model to the following sample data:**
-
● X = [1, 2, 3, 4, 5]
-
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
-
**Plot the regression line over the data points.**
-
**Answer:**
-
Simple Linear Regression is used to study the relationship between one independent variable (X) and one dependent variable (Y). In this question, we fit a regression line to the given dataset and then visualize how well the line represents the underlying trend in the data.

The goal is to learn the best-fitting straight line using the least-squares method and then plot this line along with the actual data points.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Training the Simple Linear Regression Model
model = LinearRegression()
model.fit(X, Y)

# Predicting Y values
Y_pred = model.predict(X)

# Plotting Data Points and Regression Line
plt.scatter(X, Y, color='blue', label="Actual Data")
plt.plot(X, Y_pred, color='red', label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```

**Explanation:**
-

The scatter plot displays the actual data points, and the red line represents the regression model's predicted values. The fitted line shows a clear upward trend, indicating a strong positive linear relationship between X and Y.
This model can now be used to estimate Y for any new value of X within the same range.

**Question 7: Fit a Multiple Linear Regression model on this sample data:**
-
● Area = [1200, 1500, 1800, 2000]
-
● Rooms = [2, 3, 3, 4]
-
● Price = [250000, 300000, 320000, 370000]
-
**Check for multicollinearity using VIF and report the results.**
-

**Answer:**
-
To understand how multiple independent variables (Area and Rooms) influence the house price, we apply a Multiple Linear Regression model. This model learns how each feature contributes to predicting the dependent variable (Price). After fitting the model, we compute the Variance Inflation Factor (VIF) to check for multicollinearity.
Multicollinearity occurs when predictors are strongly correlated, reducing the reliability of the model coefficients.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Sample Data
data = pd.DataFrame({
    "Area": [1200, 1500, 1800, 2000],
    "Rooms": [2, 3, 3, 4],
    "Price": [250000, 300000, 320000, 370000]
})

# Independent and Dependent Variables
X = data[["Area", "Rooms"]]
y = data["Price"]

# Fit Multiple Linear Regression Model
model = LinearRegression().fit(X, y)

# Calculate VIF for each feature
vif = pd.DataFrame()
vif["Feature"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif)
```

**Interpretation:**
-

The VIF values for Area (18.75) and Rooms (16.66) are significantly higher than the commonly accepted threshold of 10.
This indicates that strong multicollinearity exists between the two features.
This is expected because houses with larger areas often tend to have more rooms, making the features correlated.

High VIF values suggest that the model coefficients may be unstable, and interpretation of individual predictors should be done cautiously.

**Question 8: Implement polynomial regression on the following data:**
-
● X = [1, 2, 3, 4, 5]
-
● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
-
**Fit a 2nd-degree polynomial and plot the resulting curve.**
-

**Answer:**
-
Polynomial Regression is used when the relationship between the independent variable (X) and the dependent variable (Y) is not linear but follows a curved pattern. By adding higher-degree terms such as
X**2
, the model can capture non-linear trends that a simple straight line cannot.

In this question, we fit a 2nd-degree polynomial model to the given data. After transforming the input features into polynomial form, we train a regression model and visualize the fitted curve.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Transforming Data into Polynomial Features (Degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fitting Polynomial Regression Model
model = LinearRegression()
model.fit(X_poly, Y)

# Predicting Values
Y_pred = model.predict(X_poly)

# Plotting
plt.scatter(X, Y, color='blue', label="Actual Data")
plt.plot(X, Y_pred, color='red', label="Polynomial Fit")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Regression (Degree 2)")
plt.legend()
plt.show()
```

**Explanation:**
-

The polynomial curve closely follows the upward trend in the data, showing that a 2nd-degree model is suitable for capturing the non-linear growth pattern present in Y as X increases.

**Question 9: Create a residuals plot for a regression model trained on this data:**
-
● X = [10, 20, 30, 40, 50]
-
● Y = [15, 35, 40, 50, 65]
-
**Assess heteroscedasticity by examining the spread of residuals.**
-
**Answer:**
-
Residual analysis is an important step in regression evaluation. Residuals represent the difference between actual values and predicted values. A residual plot helps determine whether the model satisfies the assumption of constant variance (homoscedasticity).

In this question, we train a simple linear regression model and then plot the residuals to observe whether their spread remains consistent.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit Model
model = LinearRegression()
model.fit(X, Y)

# Predictions
Y_pred = model.predict(X)
residuals = Y - Y_pred

# Plotting Residuals
plt.scatter(X, residuals, color='green')
plt.axhline(0, color='black')
plt.xlabel("X")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

**Assessment:**
-

If the residuals are evenly scattered around the horizontal line (y = 0) with no visible pattern, the model satisfies homoscedasticity.
However, if the residual spread increases or decreases with X, or shows a clear shape, this indicates heteroscedasticity, meaning the model may not be capturing variance properly.

The goal is to have residuals that look random, suggesting a well-fitted model.

**Question 10: Addressing heteroscedasticity and multicollinearity in a real-estate regression model.**
-

**Answer:**
-

In a real-estate price prediction model, features such as area, number of rooms, and location often influence the final outcome. While building such a model, two major issues may arise: heteroscedasticity and multicollinearity. These issues negatively affect prediction accuracy, interpretation, and overall model robustness.

**To ensure a reliable model, the following steps should be taken:**

**1. Handling Heteroscedasticity**
-

● Apply Transformations:
Transforming the dependent variable (e.g., using log(price) or √price) often stabilizes variance and improves model performance.


● Use Weighted Least Squares (WLS):
This technique assigns weights to data points such that points with higher variance get lower weight, leading to stable estimates.


● Add Missing Predictors:
Sometimes uneven variance is caused by missing important features such as location quality or property age. Including such variables reduces residual patterns.


● Treat Outliers:
Extreme values can distort the regression and create heteroscedasticity. Removing or adjusting them helps maintain consistency.


**2. Handling Multicollinearity**
-

● Remove Highly Correlated Features:
If two features represent similar information (e.g., area and number of rooms), removing one reduces redundancy and improves interpretability.

● Combine Features:
Sometimes creating a combined variable, such as "area per room", solves multicollinearity while preserving information.

● Use Regularization (Ridge/Lasso):
Regularization techniques shrink coefficients and make the model more stable under multicollinearity.

● Apply PCA (Principal Component Analysis):
PCA reduces correlated variables into uncorrelated components, simplifying the model.

**Outcome :**

By addressing both heteroscedasticity and multicollinearity, the model becomes more accurate, stable, and interpretable. The predictions become trustworthy, and the regression coefficients reflect the real influence of each feature, allowing better decision-making for house price predictions.
