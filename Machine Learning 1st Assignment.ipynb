**Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent
them?**
-

**Answer:**
-
**Overfitting and Underfitting in Machine Learning**
-

**1Ô∏è. Overfitting (Model is too smart for the training data)**
-

- Overfitting happens when your model learns the training data too well ‚Äî even the noise and mistakes.
Because of this, it performs very well on training data, but poorly on new/unseen data.

- Example:
A student memorizes every question from last year‚Äôs exam. If a new type of question comes, they fail.

**Signs of Overfitting:**
-

- Training accuracy is very high

- Test/validation accuracy is low

**2Ô∏è. Underfitting (Model is not learning enough)**
-

- Underfitting happens when your model is too simple and cannot learn patterns from the data.
- So it performs poorly on both training and test data.

- Example:
A student who didn‚Äôt study at all. They don‚Äôt know old or new questions.

**Signs of Underfitting:**
-

- Training accuracy is low

- Test accuracy is also low

**How to Detect Overfitting & Underfitting**
-
**Overfitting detection:**
-

- Big gap between training and validation/test performance.

- Training accuracy ‚Üë
- Validation accuracy ‚Üì

**Underfitting detection:**
-

- Both training and validation accuracy are low.

- Model is too simple.

**How to Prevent Them**
-
**Prevent Overfitting**
-

- Cross-validation (like k-fold CV)
Helps check how well the model performs on different subsets of data.

- Regularization (L1, L2)
Adds a penalty to large weights; keeps the model from becoming too complex.

- Early stopping
Stop training when the model starts overfitting.

- More data or data augmentation
Helps model learn general patterns.

- Dropout (in deep learning)
Randomly drops neurons so they don‚Äôt rely too much on specific patterns.

**Prevent Underfitting**
-

- Use a more complex model
E.g., from linear regression ‚Üí decision tree.

- Train for longer
Increase number of epochs/iterations.

- Reduce regularization
Too much regularization makes model too simple.

- Add more useful features
Better input helps the model learn more.

**Question 3:How would you handle missing values in a dataset? Explain at least three
methods with examples.**
-
**Answer:**
-
**How to Handle Missing Values in a Dataset**
-
- Missing values are common in real datasets, and handling them properly is important because they affect model accuracy.
Here are three simple methods with examples:

**1Ô∏è‚É£ Deletion (Removing rows or columns)**
-
**a) Remove rows**
-
- If only a few rows have missing values, we can delete those rows.

- Example:
If 3 out of 1,000 rows have a missing ‚ÄúAge‚Äù value ‚Üí we can remove those 3 rows safely.

**Use when:**

- Missing data is very small

- Data removed won‚Äôt affect results

**b) Remove columns**
-
- If a whole column has too many missing values (80‚Äì90%), we delete that column.

- Example:
A ‚ÄúSalary_2020‚Äù column has 90% missing values ‚Üí better to remove it.

**2Ô∏è‚É£ Mean/Median Imputation (Simple replacement)**
-

You replace missing values with a mean, median, or mode of that column.

**üîπ Mean Imputation (for numerical data)**
-

Example:
Ages = [20, 22, 25, NaN, 30]
Mean = (20+22+25+30)/4 = 24.25
Missing age ‚Üí replaced with 24.25

**üîπ Median Imputation (better when data has outliers)**
-

Example:
Income values: [20k, 22k, 21k, 10 lakh, NaN]
Mean will be wrongly affected by 10 lakh.
So use median, which is stable.

**üîπ Mode Imputation (for categorical data)**
-

Example:
City = [Delhi, Bangalore, NaN, Delhi]
Mode = Delhi
Missing ‚Üí Delhi

**3Ô∏è‚É£ Predictive Modeling (Advanced method)**
-

Here we treat the missing value as something we predict using ML models.

**How it works:**

- Split data into two parts: rows with values and rows with missing values

- Train a model (e.g., regression, KNN, random forest) to predict the missing value

Example:
Missing "Age" can be predicted using:

- Height

- Weight

- Occupation

- Education

The model learns relationships and predicts the missing age more accurately than mean/median.

**Use when:**

- Accurate imputation is needed

- Missing values are many

- Dataset is important for predictions

**Question 4:What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).**
-

**Answer:**
-
**Imbalanced Dataset :**
-

- An imbalanced dataset is when one class has a lot of samples and the other class has very few samples.

- Because of this imbalance, the model learns more from the majority class and performs poorly on the minority class.

- Example: Fraud = 1%, Not Fraud = 99%

**Techniques to Handle Imbalanced Datasets**
-
**1. Random Oversampling / Random Undersampling**
-
**Random Oversampling :**
-

- **Theory :** Increase the size of the minority class by duplicating its samples, so both classes become balanced.

- **Practical :**
If Fraud = 200 and Not Fraud = 10,000 ‚Üí
we duplicate the Fraud samples until Fraud ‚âà 10,000.

- This helps the model learn the minority class better.

**Random Undersampling :**
-

- **Theory :** Reduce the number of samples in the majority class so that it matches the minority class size.

- **Practical :**
If Not Fraud = 10,000 and Fraud = 200 ‚Üí
reduce Not Fraud down to 200 samples.

- This is simple and fast but can remove useful information.

**2. SMOTE (Synthetic Minority Oversampling Technique)**
-

**SMOTE is an advanced oversampling method that creates synthetic samples instead of copying existing ones.**

- **Theory :**
SMOTE chooses a minority sample, finds its nearest neighbors, and creates new points in between.
This generates realistic new samples and reduces overfitting.

- **Practical :**
Suppose Fraud = 200.
SMOTE can create 800 synthetic samples ‚Üí making Fraud = 1000.
These new points help the model understand the minority class more accurately.

- SMOTE is widely used because it keeps the dataset balanced without losing information and without copying the same point again and again.

**Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?**
-
**Answer:**
-
**Comparing Label Encoding and One-Hot Encoding:**
-
**Label Encoding :**
-

- Label Encoding converts each category into a number (0, 1, 2, 3 ‚Ä¶).

- It is simple and keeps the column as a single feature.

- But it creates an order/priority between categories (like 0 < 1 < 2), which may be wrong for some data.

- Example:

  Colors = Red, Blue, Green
  Label Encoded ‚Üí Red = 0, Blue = 1, Green = 2

- Use Label Encoding when categories have a natural order (Ordinal Data).

  Examples:

   - Small < Medium < Large

   - Low < Medium < High

   - Education Level: Primary < Secondary < Graduate

**One-Hot Encoding :**
-

- One-Hot Encoding creates separate columns for each category.

- Each row gets 1 for its category and 0 for others.

= It removes any false sense of ordering.

- Example:

  Colors = Red, Blue, Green
  
  One-Hot ‚Üí

  - Red ‚Üí [1, 0, 0]

  - Blue ‚Üí [0, 1, 0]

   - Green ‚Üí [0, 0, 1]

- **Use One-Hot Encoding when categories have NO natural order (Nominal Data).**

  Examples:

  - City names

  - Gender

  - Country

  - Product type

**When to prefer which?**
-

- **Label Encoding ‚Üí When categories have order.**

  - Example: Low < Medium < High

  - Model can understand the ranking.

- **One-Hot Encoding ‚Üí When categories have no order.**

  - Example: Delhi, Mumbai, Chennai

  - Prevents the model from thinking Delhi < Mumbai < Chennai.

-

- Different app categories receive different ratings depending on app quality, complexity, user expectations, and usage type.

- Categories that are simple, focused, and useful often receive higher ratings.

- Categories with technical issues, bugs, or high user complaints often get lower ratings.



```python
import pandas as pd

# Load dataset
df = pd.read_csv("googleplaystore.csv")

# Clean rating column
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df = df.dropna(subset=['Rating', 'Category'])

# Group by category and calculate average ratings
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

print("Average Ratings by Category:\n")
print(category_ratings)

# Highest & lowest rating categories
highest = category_ratings.head(5)
lowest = category_ratings.tail(5)

print("\nTop 5 Highest Rated Categories:")
print(highest)

print("\nTop 5 Lowest Rated Categories:")
print(lowest)
```
